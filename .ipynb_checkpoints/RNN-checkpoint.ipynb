```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('cansim-0800020-eng-6674700030567901031.csv',
              skiprows=6,skipfooter=9,engine='python')
```

```python
df.head(2)
```

```python
from pandas.tseries.offsets import MonthEnd
```

```python
df['Adjustments']=pd.to_datetime(df['Adjustments'])+MonthEnd(1)
df=df.set_index('Adjustments')
```

```python
df.plot()
```

```python
split_date=pd.Timestamp('01-01-2011')
train=df.loc[:split_date,['Unadjusted']]
test=df.loc[split_date:,['Unadjusted']]
```

```python
ax=train.plot()
test.plot(ax=ax)
plt.legend(['train','test'])
```

```python
from sklearn.preprocessing import MinMaxScaler

sc=MinMaxScaler()
train_sc=sc.fit_transform(train)
test_sc=sc.transform(test)
```

```python
X_train=train_sc[:-1]
y_train=train_sc[1:]

X_test=test_sc[:-1]
y_test=test_sc[1:]
```

```python
from keras.models import Sequential
from keras.layers import Dense
import keras.backend as K
from keras.callbacks import EarlyStopping
```

```python
K.clear_session()

model=Sequential()
model.add(Dense(12,input_dim=1,activation='relu'))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
model.summary()
```

```python
early_stop=EarlyStopping(monitor='loss',patience=1,verbose=1)
```

```python
model.fit(X_train,y_train,epochs=200,
         batch_size=2,verbose=1,callbacks=[early_stop])
```

```python
y_pred=model.predict(X_test)
```

```python
plt.plot(y_test)
plt.plot(y_pred)
```

Inference from fully connected model- It doesn't predict very well for time series data

```python
X_train_t=X_train[:,None]
X_test_t=X_test[:,None]
```

```python
K.clear_session()
model=Sequential()

model.add(LSTM(6,input_shape=(1,1)))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.fit(X_train_t,y_train,epochs=150,batch_size=2,verbose=1,callbacks=[early_stop])
```

```python
y_pred=model.predict(X_test_t)
plt.plot(y_test)
plt.plot(y_pred)
```
